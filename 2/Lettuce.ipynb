```python
import numpy as np
import pandas 
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

## U datasetu dati su sledeci parametri:
## 1. Plant Id- identifikacija svake biljke
## 2. Date- datum
## 3. Temperature- temperatura data u Celzijusovoj skali
## 4. Humidity - procenat vlage u vazduhu
## 5. Total Dissolved Solids (TDS) Value- Ukupne rastvorene cvrste materija
## 6. ph levels- ph vrednosti
## 7. Growth days- broj dana potreban da biljka dostigne zrelost

```python
lettuce_df= pandas.read_csv("C:/Users/Kristina/Desktop/Kupus/lettuce_dataset.csv",encoding='latin-1')
lettuce_df.head(10)
```

```python
lettuce_df.tail(10)
```

## Distribucija podataka o temperaturi

```python
sns.histplot(data=lettuce_df, x=lettuce_df["Temperature (°C)"], kde=True, color="blue")
```

## Distribucija podataka o vlazi

```python
sns.histplot(data=lettuce_df, x=lettuce_df["Humidity (%)"], kde=True, color="red")
```

## Distribucija podataka o TDS

```python
sns.histplot(data=lettuce_df, x=lettuce_df["TDS Value (ppm)"], kde=True, color="green")
```

## Distribucija podataka o danima rasta

```python
sns.histplot(data=lettuce_df, x=lettuce_df["Growth Days"], kde=True, color="brown")
```

## Distribucija podataka o pH vrednostima

```python
sns.histplot(data=lettuce_df, x=lettuce_df["pH Level"], kde=True, color="purple")
```

## Temperatura i dani rasta

```python
fig = px.scatter(lettuce_df, x="Temperature (°C)", y="Growth Days", marginal_x="histogram", marginal_y="histogram",title="Growth Days Vs Temperature", color_discrete_sequence=["red"])
fig.show()
```

### Zakljucujemo da temperatura varira od 18 do 25 stepeni. Vidimo porciju u kojoj se temperatura krece od 29 do 34 stepeni, gde zakljucujemo da kupus nije mogao da podnese visoke temperature.

## pH vrednost i dani rasta

```python
fig = px.box(lettuce_df, x="pH Level", y="Growth Days",title="Growth Days Vs pH Level", color_discrete_sequence=["blue"])
fig.show()
```

### Zakljucujemo da drugacije pH vrednosti nemaju veliki uticaj na rast.
